```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_linear_regression

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense, Dropout
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import MeanSquaredError
```

```python
# Creating data

X_train = np.linspace(-1, 1, 20)

X_test = np.linspace(-1, 1, 20)

y_train = np.array([-0.6561 , -0.3099 , -0.59035, -0.50855, -0.285  , 
                    -0.2443 , -0.02445,  0.00135, -0.2006 ,  0.07475, 
                    -0.1422 ,  0.06515,  0.15265,  0.3521 ,  0.28415,  
                    0.5524 ,  0.23115,  0.20835, 0.4211,  0.60485])

y_test = np.array([-0.69415, -0.451  , -0.43005, -0.4484 , -0.1475 ,
                   -0.5019 , -0.28055,  0.24595, -0.21425, -0.0286 ,  
                   0.23415,  0.46575, 0.07955,  0.1973 ,  0.0719 ,
                   0.3639 ,  0.5536 ,  0.3365 , 0.50705,  0.33435])
```

```python
X_train.shape
```

```python
plt.scatter(X_train, y_train, color="red", label="training points")
plt.scatter(X_test, y_test, color="black", label="test points")
plt.legend()
plt.show()
```

## Without Dropouts

```python
# creating regression model

model = Sequential()

model.add(InputLayer(shape=(1, )))
model.add(Dense(128, activation="relu"))
model.add(Dense(128, activation="relu"))
model.add(Dense(1, activation="linear"))

adam = Adam(learning_rate=0.01)

# model.compile(loss="mse", optimizer=adam, metrics=["r2_score"])    
model.compile(loss=MeanAbsoluteError(), optimizer=adam, metrics=["r2_score"])    # just trying out new losses
```

```python
history = model.fit(X_train, y_train, epochs=500, validation_data=(X_test, y_test))
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 5))

ax[0].plot(history.history["loss"], label="train loss")
ax[0].plot(history.history["val_loss"], label="validation loss")

ax[1].plot(history.history["r2_score"], label="train r2 score")
ax[1].plot(history.history["val_r2_score"], label="validation r2 score")

ax[0].legend()
ax[1].legend()

plt.show()
```

```python
y_pred = model.predict(X_test)

plt.scatter(X_train, y_train, c='red', label='Train')
plt.scatter(X_test, y_test, c='black', label='Test')
plt.plot(X_test, y_pred)
plt.legend()
plt.ylim((-1.5, 1.5))
plt.show()
```

```python
# We can clearly see that the model is overfitting
# Now let us try the same model with dropout layers
```

## With Dropout Layers

```python
model_2 = Sequential()

model_2.add(InputLayer(shape=(1, )))
model_2.add(Dense(32, activation="relu"))    # hidden layer 1
model_2.add(Dropout(0.2))                    # dropout layer for hidden layer 1
model_2.add(Dense(32, activation="relu"))    # hidden layer 2
model_2.add(Dropout(0.2))                    # dropout layer for hidden layer 2
model_2.add(Dense(1, activation="linear"))

adam = Adam(learning_rate=0.01)

model_2.compile(loss="mae", optimizer=adam, metrics=["r2_score"])
```

```python
history_2 = model_2.fit(X_train, y_train, epochs=500, validation_data=(X_test, y_test))
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 5))

ax[0].plot(history.history["loss"], label="train loss")
ax[0].plot(history.history["val_loss"], label="validation loss")
ax[0].legend()
ax[0].set_title("Without Dropout Layer")

ax[1].plot(history_2.history["loss"], label="train loss with p=0.2")
ax[1].plot(history_2.history["val_loss"], label="validation loss with p=0.2")
ax[1].legend()
ax[1].set_title("With Dropout Layer")

plt.show()
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 5))

ax[0].plot(history.history["r2_score"], label="train r2_score")
ax[0].plot(history.history["val_r2_score"], label="validation r2_score")
ax[0].legend()
ax[0].set_title("Without Dropout Layer")

ax[1].plot(history_2.history["r2_score"], label="train r2_score with p=0.2")
ax[1].plot(history_2.history["val_r2_score"], label="validation r2_score with p=0.2")
ax[1].legend()
ax[1].set_title("With Dropout Layer")

plt.show()
```

```python
# We can see that validation loss decreased and the model is now no longer overfitting as much
# Same goes for r2 score  

# However the training loss increased a bit because the model has reduced bias now
```

## Effect of different values of p

```python
model_3 = Sequential()

model_3.add(InputLayer(shape=(1, )))
model_3.add(Dense(32, activation="relu"))    
model_3.add(Dropout(0.75))                    
model_3.add(Dense(32, activation="relu"))    
model_3.add(Dropout(0.75))                    
model_3.add(Dense(1, activation="linear"))

adam = Adam(learning_rate=0.01)

model_3.compile(loss="mae", optimizer=adam, metrics=["r2_score"])
```

```python
history_3 = model_3.fit(X_train, y_train, epochs=500, validation_data=(X_test, y_test), verbose=0)
```

```python
y_pred = [0] * 3
y_pred[0] = model.predict(X_test)
y_pred[1] = model_2.predict(X_test)
y_pred[2] = model_3.predict(X_test)

fig, ax = plt.subplots(ncols=3, figsize=(15, 5))

for i in range(3):
    ax[i].scatter(X_train, y_train, c='red', label='Train')
    ax[i].scatter(X_test, y_test, c='black', label='Test')
    ax[i].plot(X_test, y_pred[i])
    ax[i].legend()
    ax[i].set_ylim((-1.5, 1.5))

ax[0].set_title("Without Dropouts")
ax[1].set_title("Without Dropouts p=0.2")
ax[2].set_title("Without Dropouts p=0.5")

plt.show()
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(15, 5))

ax[0].plot(history.history["loss"], label="train loss")
ax[0].plot(history.history["val_loss"], label="validation loss")
ax[0].legend()
ax[0].set_title("Without Dropout Layer")

ax[1].plot(history_2.history["loss"], label="train loss with p=0.2")
ax[1].plot(history_2.history["val_loss"], label="validation loss with p=0.2")
ax[1].legend()
ax[1].set_title("With Dropout Layer")

ax[2].plot(history_3.history["loss"], label="train loss with p=0.75")
ax[2].plot(history_3.history["val_loss"], label="validation loss with p=0.75")
ax[2].legend()
ax[2].set_title("With Dropout Layer")

plt.show()
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(15, 5))

ax[0].plot(history.history["r2_score"], label="train r2_score")
ax[0].plot(history.history["val_r2_score"], label="validation r2_score")
ax[0].legend()
ax[0].set_title("Without Dropout Layer")

ax[1].plot(history_2.history["r2_score"], label="train r2_score with p=0.2")
ax[1].plot(history_2.history["val_r2_score"], label="validation r2_score with p=0.2")
ax[1].legend()
ax[1].set_title("With Dropout Layer p=0.2")

ax[2].plot(history_3.history["r2_score"], label="train r2_score with p=0.75")
ax[2].plot(history_3.history["val_r2_score"], label="validation r2_score with p=0.75")
ax[2].legend()
ax[2].set_title("With Dropout Layer p=0.75")

plt.show()
```

```python
# As we can see too too large of a dropout ratio will cause underfitting
# Generally its kept between 0.1 and 0.5 for ANN
# 0.4 - 0.5 for CNNs
# 0.2 - 0.3 for RNNs

# But its upto experimentation and finding the best value no hard rule
```

## Genreally dropouts are only to the last hidden layer. Let's see what happens there 
- No steadfast rule here also its just generally done like this. If we don't see improvement we can always add more dropout layers

```python
model_4 = Sequential()

model_4.add(InputLayer(shape=(1, )))
model_4.add(Dense(32, activation="relu"))    # hidden layer 1                    
model_4.add(Dense(32, activation="relu"))    # hidden layer 2
model_4.add(Dropout(0.5))                    # dropout layer only added to hidden layer 2
model_4.add(Dense(1, activation="linear"))

adam = Adam(learning_rate=0.01)

model_4.compile(loss="mae", optimizer=adam, metrics=["r2_score"])
history_4 = model_4.fit(X_train, y_train, epochs=500, validation_data=(X_test, y_test), verbose=0)
```

```python
fig, ax = plt.subplots(ncols=3, figsize=(15, 5))

ax[0].plot(history_4.history["loss"], label="train loss")
ax[0].plot(history_4.history["val_loss"], label="validation loss")
ax[0].legend()
ax[0].set_title("loss")

ax[1].plot(history_4.history["r2_score"], label="train r2_score")
ax[1].plot(history_4.history["val_r2_score"], label="validation r2_score")
ax[1].legend()
ax[1].set_title("r2 score")

y_pred_4 = model_4.predict(X_test)

ax[2].scatter(X_train, y_train, c='red', label='Train')
ax[2].scatter(X_test, y_test, c='black', label='Test')
ax[2].plot(X_test, y_pred_4)
ax[2].legend()
ax[2].set_ylim((-1.5, 1.5))

plt.show()
```
