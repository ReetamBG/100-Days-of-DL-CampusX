```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import InputLayer, Dense
```

```python
df = pd.read_csv("./datasets/Social_Network_Ads.csv")
df.head()
```

```python
X = df.drop(columns="Purchased")
y = df["Purchased"]
```

## Without Feature Scaling

```python
sns.scatterplot(data=df, x="Age", y="EstimatedSalary")
```

```python
model = Sequential()

model.add(InputLayer(shape=(2, )))
model.add(Dense(32, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.summary()
```

```python
model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
history = model.fit(X, y, validation_split=0.2, epochs=100)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

sns.lineplot(history.history["loss"], label="train loss", ax=ax[0])
sns.lineplot(history.history["val_loss"], label="validation loss", ax=ax[0])
ax[0].legend()
ax[0].set_ylim(0, 500)

sns.lineplot(history.history["accuracy"], label="train accuracy", ax=ax[1])
sns.lineplot(history.history["val_accuracy"], label="validation accuracy", ax=ax[1])
ax[1].legend()

plt.show()
```

```python
# As we can see the training is very janky and does not converge properly
# The model is not even being able to be trained properly
```

```python
plot_decision_regions(X=X.values, y=y.values, clf=model)
```

## With Feature Scaling

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_trf = scaler.fit_transform(X)

X_trf[10], X.values[10]
```

```python
fig, ax = plt.subplots(ncols=2, figsize=(15, 5))

sns.scatterplot(x=X.values[:, 0], y=X.values[:, 1], ax=ax[0])
ax[0].set_title("Before Scaling")
sns.scatterplot(x=X_trf[:, 0], y=X_trf[:, 1], ax=ax[1])
ax[1].set_title("After Scaling")
```

```python
model_2 = Sequential()

model_2.add(InputLayer(shape=(2, )))
model_2.add(Dense(32, "relu"))
model_2.add(Dense(1, "sigmoid"))

model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
history_2 = model.fit(X_trf, y, epochs=100, validation_split=0.2)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(15, 5))

sns.lineplot(history_2.history["loss"], label="train loss", ax=ax[0])
sns.lineplot(history_2.history["val_loss"], label="validation loss", ax=ax[0])
ax[0].legend()
# ax[0].set_ylim(0, 500)

sns.lineplot(history_2.history["accuracy"], label="train accuracy", ax=ax[1])
sns.lineplot(history_2.history["val_accuracy"], label="validation accuracy", ax=ax[1])
ax[1].legend()

plt.show()
```

```python
# Now it is more streamlined and stable and the model is being trained atleast
```
