```python
import numpy as np
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import make_classification
```

```python
X, y = make_classification(n_samples=100,
                           n_features=2, 
                           n_informative=1,
                           n_redundant=0,
                           n_classes=2,
                           n_clusters_per_class=1, 
                           random_state=41,
                           hypercube=False,
                           class_sep=10)

plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
class Perceptron():
    def __init__(self, epochs=1000, lr=0.1):
        self.epochs = epochs
        self.lr = lr
        self.W = None    # weights 

    def step(self, x):
        return 0 if x<0 else 1

    def fit(self, X, y):
        n_rows = X.shape[0]
        n_cols = X.shape[1]
        
        self.W = np.ones(n_cols + 1)    # added +1 due to W0 for X0 which is added in the line below
        X = np.insert(X, 0, 1, axis=1)
        
        for epoch in range(self.epochs):
            random_idx = np.random.randint(0, n_rows)
            y_pred = self.step(self.W @ X[random_idx])
            
            self.W = self.W + self.lr * (y[random_idx] - y_pred) * X[random_idx]

        intercept_ = self.W[0]
        coef_ = self.W[1:]

        return intercept_, coef_

    def predict(self, X):
        X = np.insert(X, 0, 1, axis=1)
        y_pred = X @ self.W
        y_pred = [self.step(pred) for pred in y_pred]
        return y_pred
```

```python
from sklearn.metrics import accuracy_score

p = Perceptron()
intercept_, coef_ = p.fit(X, y)
m = -(coef_[0]/coef_[1])
b = -(intercept_/coef_[1])
X_plot = np.linspace(-3,3,100)
y_plot = m*X_plot + b

y_pred = p.predict(X)
print("ACCURACY SCORE :", accuracy_score(y, y_pred))

plt.figure(figsize=(15,5))
plt.plot(X_plot,y_plot,color='darkred',linewidth=3)
plt.scatter(X[:,0], X[:,1], c=y)
plt.ylim(-3,2)

plt.show()
```
