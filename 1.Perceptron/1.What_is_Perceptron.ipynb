```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron
```

```python
data = load_iris()
X = data.data[:, 0:2]    # only petal length and sepal length is taken
y = data.target
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
p = Perceptron()
p.fit(X, y)
```

```python
# Perceptron is only for binary classification don't forget
```

```python
plot_decision_regions(X=X, y=y, clf=p)
```
